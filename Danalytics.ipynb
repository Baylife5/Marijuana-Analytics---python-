```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as py
from numpy.core.fromnumeric import size
import pygame
```

```python
hybrids ={}

size = input ("how many stains would you need to input?")
        
for i in range (int(size)):       
 key = input("Enter strian") 
 value = input ("what is the thc count? ")
 hybrids[key] = value


print(hybrids)
```

```python
l = []
s = []

for a in hybrids:
     l.append(a)
     s.append(int(hybrids[a]))

''' if we want the list sorted'''
#l = sorted(l)
#s = sorted(s)
py.bar(l ,s, color = "green")
#py.plot(l, s, marker = '*', color = 'blue')
py.title("THC count/balance")
py.show()
```
